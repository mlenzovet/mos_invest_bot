```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage to scrape
url = "https://investmoscow.ru/about-moscow/project-details/?project=1"

# Send a GET request to the URL
response = requests.get(url)

# Parse the webpage content
soup = BeautifulSoup(response.content, "html.parser")

# Define the structure to extract the required data
project_data = []

# Extract the project description block
project_description = soup.find("div", class_="projects__article")

# Check if the project description block is found
if project_description:
    # Extract the title
    title = project_description.find("h4").get_text(strip=True)
    
    # Extract the paragraphs within the description
    description_paragraphs = project_description.find_all("p")
    description_text = "\n".join([p.get_text(strip=True) for p in description_paragraphs])

    # Extract financing details
    financing_detail = project_description.find_all("div", class_="row py-4")[0].find_all("p")[1].get_text(strip=True)

    # Extract implementation duration
    implementation_duration = project_description.find_all("div", class_="row py-4")[1].find_all("p")[1].get_text(strip=True)

    # Extract current status
    current_status = project_description.find_all("div", class_="row py-4")[2].find_all("p")[1].get_text(strip=True)

    # Append the data to the list
    project_data.append([title, description_text, financing_detail, implementation_duration, current_status])

# Create a DataFrame
df = pd.DataFrame(project_data, columns=["Title", "Description", "Financing Detail", "Implementation Duration", "Current Status"])

# Display the DataFrame to the user
df
```
